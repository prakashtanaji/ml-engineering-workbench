```python
import numpy as np
# import matplotlib.pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D
```

Generate inputs:

```python
observations = 1000

#two variable linear model
xs = np.random.uniform(low=-10, high=10,size=(observations,1))
zs = np.random.uniform(low=-10, high=10,size=(observations,1))
#print(xs,zs)

inputs = np.column_stack((xs,zs))
print(inputs.shape)
```

Create targets:

```python
#targets = f(x,z) = 2x - 3z +5 + noise
noise = np.random.uniform(-1,1,size=(observations,1))
targets = 2*xs - 3*zs + 5 + noise

print(targets.shape)
```

Plot results:

Initialize variables:

```python
init_range = 0.1
weights = np.random.uniform(-init_range, init_range,size=(2,1))
biases = np.random.uniform(-init_range, init_range,size=1)

print(weights, biases)

print(weights.shape, biases.shape)
```

Learning rate:

```python
learning_rate = 0.001
```

Training:

    Plan:
        - calculate outputs
        - compare outputs to targets through loss
        - print loss
        - Adjust weights and biases
        - Repeat ~

```python
for i in range(10000):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - targets

    #L2-norm loss formula -> Ei(yi-ti)squared 2
    loss = np.sum(deltas**2)/2/observations

    print(loss)

    deltas_scaled=deltas / observations

    #print(delta_scaled.shape)

    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
```

```python
print(weights, biases)
```
